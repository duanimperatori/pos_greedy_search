```python
def maximizar_pontos_sobrevivencia(capacidade_mochila, pesos, pontos):
    n = len(pesos)

    # Inicializando a tabela de programação dinâmica para armazenar os resultados intermediários
    dp = [[0 for _ in range(capacidade_mochila + 1)] for _ in range(n + 1)]

    # Preenchendo a tabela de programação dinâmica
    for i in range(1, n + 1):
        for w in range(1, capacidade_mochila + 1):
            print(f'i: {i} | w: {w} | item: {i} | pontos: {pontos[i - 1]} | peso: {pesos[i - 1]} | ' +
                  f'comp1: {dp[i - 1][w]} | comp2: {pontos[i - 1] + dp[i - 1][w - pesos[i - 1]]} | range: {[i - 1] + [w - pesos[i - 1]]} | ' +
                  f'max: {max(dp[i - 1][w], pontos[i - 1] + dp[i - 1][w - pesos[i - 1]])}')
            if pesos[i - 1] > w:
                dp[i][w] = dp[i - 1][w]
            else:
                # Maximizando entre escolher ou não escolher o item
                dp[i][w] = max(dp[i - 1][w], pontos[i - 1] + dp[i - 1][w - pesos[i - 1]])

    # Recuperando os itens escolhidos
    itens_escolhidos = []
    i, w = n, capacidade_mochila
    while i > 0 and w > 0:
        if dp[i][w] != dp[i - 1][w]:
            itens_escolhidos.append(i - 1)
            w -= pesos[i - 1]
        i -= 1

    # Retornando os pontos máximos e os itens escolhidos
    return dp[n][capacidade_mochila], itens_escolhidos, dp

# Dados do problema
capacidade_mochila = 30
nome_item = ['Saco Dormir', 'Corda', 'Canivete', 'Tocha', 'Garrafa', 'Comida']
pesos = [15, 3, 2, 5, 9, 20]
pontos = [15, 7, 10, 5, 8, 17]

# Chamando a função
pontos_maximos, itens_escolhidos, df = maximizar_pontos_sobrevivencia(capacidade_mochila, pesos, pontos)

# Exibindo resultados
print(f"Pontos máximos de sobrevivência: {pontos_maximos}")
print("Itens escolhidos:")
for item in itens_escolhidos:
    print(f"{nome_item[item]}: Peso = {pesos[item]}, Pontos = {pontos[item]}")
```

```python
from itertools import combinations

peso_mochila = 30
items = ['Saco Dormir', 'Corda', 'Canivete', 'Tocha', 'Garrafa', 'Comida']
pesos = [15, 3, 2, 5, 9, 20]
pontos = [15, 7, 10, 5, 8, 17]
n_items = len(items)
maior_pontuacao = 0
maior_peso = 0
melhor_combinacao = []

for i in range(1, n_items + 1):
    
    combination = combinations(list(range(1, 7)), i)
    for c in combination:
        sum_pontos = 0
        sum_peso = 0
        for item in c:
            sum_pontos += pontos[item - 1]
            sum_peso += pesos[item - 1]
            print(f'item: {items[item - 1]} -- peso: {pesos[item - 1]} -- pontos: {pontos[item - 1]}')
        
        if sum_peso <= peso_mochila:
            if sum_pontos > maior_pontuacao:
                maior_pontuacao = sum_pontos
                maior_peso = sum_peso
                melhor_combinacao = c

        print(f'total peso: {sum_peso}')
        print(f'total pontos: {sum_pontos}')
        print('\n')
    

print('=' * 100)
print(' -- Melhor combinacao')
print('=' * 100)

for item in melhor_combinacao:
    sum_pontos += pontos[item - 1]
    sum_peso += pesos[item - 1]
    print(f'item: {items[item - 1]} -- peso: {pesos[item - 1]} -- pontos: {pontos[item - 1]}')
print(f'total peso: {maior_peso}')
print(f'total pontos: {maior_pontuacao}')
print('=' * 100)
```
